```python
import matplotlib.pyplot as plt
```

## Line Plot



```python
x = [1,2,3,4,5,6]
y = [1,8,1,8,1,8]
plt.plot(x, y)
# data notebook systems will automatically call plt.show() at the end of each Python script.
# 所以下面的可以忽略
plt.show()
```

```python
plt.plot(x, y, color='red')
```

## Bar chart


```python
a = ['A','B','C']
b = [4, 1, 2]
plt.bar(a, b)
```

### Horizontal bar chart


```python
a = ['A','B','C']
b = [4, 1, 2]
plt.barh(a, b)
```

## Scatter plot


```python
a = [1, 2, 3, 4]
b = [1, 4, 3, 4]
plt.scatter(a, b)
```
