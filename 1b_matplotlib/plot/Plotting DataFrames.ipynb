```python
Matplotlib integrates seamlessly with Pandas, making it straightforward to turn dataframes into visualizations.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('earth-layers.csv')
df #display it

```

```python
x = df['layer']
y = df['thickness']
plt.bar(x, y)
```

```python

plt.bar( df['layer'], df['thickness'] )
plt.ylabel('Thickness (km)')
plt.title('Layers of the Earth')
```
